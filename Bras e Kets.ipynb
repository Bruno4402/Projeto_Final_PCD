<h2>Calculador de Bras e Kets</h2>

<h5>Antes de iniciar com as operações, defino uma função limpar para simplificar visualmente o resultado final das operações.</h5>

```python
def limpar(x):
    """Simplifica visualmente o resultado de uma operação, eliminando o zero da parte real ou imaginária, quando existir.
    
    Entrada: uma matriz (list) ou um número (int, float, complex).
    Saída: retorna o resultado simplificado:
           - retorna int se a parte imaginária for zero e a real for inteira;
           - retorna complex se ambas as partes existirem;
           - retorna list, caso x seja uma matriz (lista).
    """
    
    # para matriz
    if isinstance(x, list):
        return [limpar(z) for z in x]

    # para número complexo
    if isinstance(x, complex):
        real = x.real
        imag = x.imag

        # parte real
        if isinstance(real, float) and real.is_integer():
            real = int(real)

        # parte imaginária
        if isinstance(imag, float) and imag.is_integer():
            imag = int(imag)

        if imag == 0:
            return real
        elif real == 0:
            return complex(0, imag)
        else:
            return complex(real, imag)

    return x
```

<h4>O que são bras e kets?</h4>

<h5>Na notação do físico Paul Dirac (conhecida como notação bra-ket), bras e kets são vetores do espaço de Hilbert ($\mathbb{C}$) representados matricialmente.</h5>
<h5>São muito utilizados em cálculos de mecânica e computação quântica.</h5>

<h4>1. Definindo funções de operações com kets</h4>

<h5>Ket é um vetor representado em uma matriz coluna:</h5>

<h4>$$\mathit{ket}{\quad}|\psi\rangle{\:}=\begin{bmatrix} a \\ b \\ c \end{bmatrix} $$</h4>

<h5>Nesse programa, o ket será representado como uma lista com cada elemento sendo uma outra lista: 
    
<h5 style="text-align: center;">$[{\,}[a],{\,} [b],{\,} [c]{\,}]$</h5>

<h5>A primeira operação com kets será a soma de dois kets.</h5>

```python
def soma_kets(k1, k2):
    """Calcula a soma de dois kets.
    
    Entrada: dois kets do tipo list, representados como: [[a], [b], [c]]
    Saída: retorna um dado (simplificado, conforme a função limpar) do tipo list sendo o resultado da soma desses dois kets.
    """
    
    ket_soma = []
    for i in range(len(k1)):
        soma = k1[i][0] + k2[i][0]
        ket_soma.append([soma])
    return limpar(ket_soma)
```

```python
soma_kets([[1], [2] , [4]], [[4], [5], [6]])
```

<h5>Em seguida, a multiplicação de um ket por um escalar, ou seja, produto por escalar.</h5>

```python
def multip_escalar(a, k1):
    """Calcula o produto de um ket por um escalar.
    
    Entrada: um escalar do tipo int, float ou complex e um ket do tipo list, representado como: [[a], [b], [c]]
    Saída: um ket do tipo list, sendo o resultado da multiplicação do escalar pelo ket.    
    """
    
    ket_produto_por_escalar = []
    for i in range(len(k1)):
        produto_por_escalar = a * k1[i][0]
        ket_produto_por_escalar.append([produto_por_escalar])
    return ket_produto_por_escalar
```

```python
multip_escalar(6j, [[1],[4],[7]])
```

<h4>2. Definindo bras</h4>

<h5>Bra é um ket transposto e conjugado, sendo representado por uma matriz linha.</h5>
<h5>A operação de transposição e conjugação de um ket para a formação de um bra é conhecida como adaga ($\dagger$).</h5>

<h4>$$\mathit{bra}{\quad}\langle\psi|=((|\psi\rangle)^T)^*=(|\psi\rangle)^\dagger$$</h4>

<h4>$$\Biggl(\Biggl(\begin{bmatrix} a + bi \\ c \end{bmatrix}\Biggr)^T\Biggr)^* = \begin{bmatrix} a - bi & c \end{bmatrix}$$</h4>

<h5>Nesse programa, o bra será representado como uma lista com uma sublista interna contendo os elementos (matriz linha):</h5>

<h5 style="text-align: center;">$[{\,}[a,{\,} b,{\,} c]{\,}]$</h5>

```python
def conjugacao(z):
    """Calcula o conjugado de um número complexo.
    
    Entrada: um número complexo do tipo complex, int ou float.
    Saída: retorna o número conjugado do tipo complex.    
    """
    
    z_conjugado = z.real - z.imag * 1j
    return z_conjugado
```

```python
    conjugacao(2+4j)
```

```python
import numpy as np

def bra(k1):
    """Calcula o bra correspondente a um ket.
    
    Entrada: um ket do tipo list, representado como: [[a], [b], [c]]
    Saída: retorna o bra correspondente, do tipo list, com os valores simplificados pela função limpar.
    """
    
    transposta = np.asarray(k1).T
    adaga = conjugacao(transposta).tolist()
    return limpar(adaga)
```

```python
bra([[3], [7j], [1j]])
```

<h4>3. Definindo operações com bras e kets</h4>

<h5>A primeira operação será o produto interno, caracterizado como a soma da multiplicação componente a componente de um bra por um ket. O resultado desta operação é um escalar.</h5>

<h4>$\mathit{produto\ interno}{\quad}\langle\psi|\phi\rangle$</h4>

```python
def produto_interno(k1, k2):
    """Calcula o produto interno a partir de dois kets; 
    o primeiro é convertido em um bra.

    Entrada: dois kets do tipo list, representados como: [[a], [b], [c]]
    Saída: retorna um escalar (simplificado, conforme a função limpar) 
           do tipo int, float ou complex, sendo o resultado do produto interno.
    """

    if len(k1) == len(k2):
        b1 = bra(k1)
        resultado = 0
        for i in range(len(b1[0])):
            resultado += b1[0][i] * k2[i][0]

        return limpar(resultado)

    else:
        raise ValueError("Os kets possuem dimensões diferentes!")
```

```python
produto_interno([[1], [4], [2j]], [[1], [3], [4]])
```

<h5>Em seguida, é importante calcular o módulo do ket, ou seja, a raiz quadrada do produto interno.</h5>

```python
import math

def modulo(k1):
    """Calcula o módulo de um ket, calculando a raiz quadrada do seu produto interno.

    Entrada: um ket do tipo list, representado como: [[a], [b], [c]]
    Saída: retorna um valor do tipo float (arredondado em duas casas decimais), 
           sendo a raiz quadrada do produto interno desse ket por seu bra correspondente.
    """
    
    p1 = produto_interno(k1, k1)  
    modulo = math.sqrt(p1) 
    return round(modulo, 2)
```

```python
modulo([[1], [4], [2j]])
```

<h5>Para encontrar o ket normalizado, divide-se o ket por seu módulo.</h5>

```python
def normalizado(k1):
    """Normaliza o ket, dividindo seus componentes pelo módulo do ket.
    
    Entrada: um ket do tipo list, representado como: [[a], [b], [c]]
    Saída: retorna um novo ket (list) contendo os componentes do ket inserido normalizados, 
           com os valores arredondados em duas casas decimais e simplificados, pela função limpar.
    """
    
    m1 = modulo(k1)
    normalizado = []
    for linha in k1:
        novo_valor = linha[0] / m1

        if isinstance(novo_valor, complex): # arredondamento
            resultado = complex(round(novo_valor.real, 2), round(novo_valor.imag, 2))
        else:
            resultado = round(novo_valor, 2)
        
        normalizado.append(resultado)

    return limpar(normalizado)
```

```python
normalizado([[1], [4], [2j]])
```

<h5>Outra operação envolvendo bras e kets é o produto externo, que consiste na multiplicação de um ket por um bra. O resultado desta operação é uma matriz.</h5>

<h4>$\mathit{produto\ externo}{\quad}|\phi\rangle\langle\psi|$</h4>

```python
def produto_externo(k1, k2):
    """Calcula o produto externo a partir de dois kets; 
    o segundo é convertido em um bra

    Entrada: dois kets do tipo list, representados como: [[a], [b], [c]]
    Saída: retorna uma matriz (list de lists) simplificada pela função limpar,
           sendo o resultado do produto do primeiro ket pelo bra do segundo.
    """

    if len(k1) == len(k2):
        b2 = bra(k2)
        resultado = []

        for linha_ket in k1:
            nova_linha_matriz = []
            valor_ket = linha_ket[0]

            for valor_bra in b2[0]:
                produto = valor_ket * valor_bra
                nova_linha_matriz.append(produto)

            resultado.append(nova_linha_matriz)

        return limpar(resultado)
            
    else:
        raise ValueError("Os kets tem dimensões diferentes!")    
```

```python
produto_externo([[1], [4], [2j]], [[1], [3], [4]])
```

<h5>Um operador é uma transformação linear cujo espaço vetorial de domínio é igual ao contradomínio: $${T}{\,}{:}{\:}{V}{\,}\rightarrow{\,}{V}$$</h5>

<h5>A partir disso, pode-se realizar uma transformação de um ket a partir de um operador:</h5>

```python
def operador_ket(k1, operador):
    """Calcula a aplicação de um operador em um ket.
    
    Entrada: um ket do tipo list, representado como: [[a], [b], [c]] 
             e um operador (matriz quadrada) do tipo list de lists de mesma dimensão.
    Saída: retorna um novo ket do tipo list (simplificado, conforme a função limpar) 
           sendo o resultado da aplicação do operador sobre o ket original.
    """
    
    n = len(k1)
    if len(operador) == n and len(operador[0]) == n:
        
        novo_ket = []
        
        for i in range(n):
            novo_valor = 0
            for j in range(n):
                novo_valor += operador[i][j] * k1[j][0]
            novo_ket.append([novo_valor])

    return limpar(novo_ket)
```

```python
operador_ket([[1], [2j]], [[1, 0], [0, -1]])
```

<h5>Pode-se também calcular o elemento de matriz, ou seja, uma operação que posiciona um operador Â entre um bra e um ket, resultando em um escalar.</h5>

<h4>$\mathit {elemento}{\:}{de}{\:}{matriz}{\quad}\langle\psi|{\:}\mathit {Â}{\:}|\phi\rangle$</h4>

```python
def elemento_matriz(k1, k2, operador):
    """Calcula o elemento de matriz.
    
    Entrada: dois kets do tipo list, representados como: [[a], [b], [c]],
             e um operador (matriz quadrada) do tipo list de lists compatível.
    Saída: retorna um escalar do tipo int, float ou complex, sendo o 
           resultado da operação que posiciona o operador entre o bra do primeiro 
           ket e o ket do segundo ket.
    """
    
    n = len(k1)
    if len(k2) == n:
        novo_ket = operador_ket(k2, operador)            
        elemento = produto_interno(k1, novo_ket)
        return elemento
       
    else:
        raise ValueError("Dimensões incompatíveis para o cálculo do elemento de matriz!")
```

```python
elemento_matriz([[1], [4], [2j]], [[1], [3], [4]], [[1, 0, 1],[0, 1, 0], [0, 0, 0]])
```

```python
elemento_matriz([[1], [-3j]], [[2j], [8]], [[1j, 0], [0, -1j]])
```

<h5>O valor esperado é um caso particular de elemento de matriz calculado utilizando o mesmo ket em ambos os lados da operação.</h5>

<h4>$\mathit {valor}{\:}{esperado}{\quad}\langle\psi|{\:}\mathit {Â}{\:}|\psi\rangle$</h4>

```python
def valor_esperado(k1, operador):
    """Calcula o valor esperado.
    
    Entrada: um ket do tipo list, representado como: [[a], [b], [c]]
             e um operador (matriz quadrada) do tipo list de lists compatível.
    Saída: retorna um escalar do tipo int, float ou complex, sendo o valor
           esperado obtido ao calcular o elemento de matriz usando o mesmo ket.
    """
    
    resultado = elemento_matriz(k1, k1, operador)
    return resultado
```

```python
valor_esperado([[1], [4], [2j]], [[1, 0, 1],[0, 1, 0], [0, 0, 0]])
```

```python
valor_esperado([[15], [-3], [4j]], [[2, 5, 7], [9, 7, 5], [6, 5j, 0]])
```

<h4>4. O calculador de bras e kets</h4>

<h5>Por fim, todas as operações definidas acimas foram incluídas em um menu, possibilitando ao usuário que ele escolha uma operação, entre com valores e receba o resultado.</h5>

```python
continuar = True

while continuar:
    entrada = input(f""" 
    Calculador de Bras e Kets
    -----------------------------------------
    Qual operação você deseja executar?
    Digite o número correspondente:
    
    1 - Somar dois kets
    2 - Multiplicar um ket por um escalar
    3 - Converter um ket em um bra
    4 - Produto interno 
    5 - Módulo
    6 - Ket normalizado
    7 - Produto externo
    8 - Aplicação de um operador em um ket
    9 - Elemento de matriz
    10 - Valor esperado
    0 - Sair do calculador
        \n""")

    try:
        operacao = int(entrada)
    except ValueError:
        operacao = -1

    if operacao == 0:
        print("\nEncerrando o calculador... Até mais!")
        continuar = False
        continue
    
    if operacao == 1:
        k1 = eval(input(f"\nDigite o primeiro ket: "))
        k2 = eval(input(f"Digite o segundo ket: "))
        print(f"A soma é: {soma_kets(k1, k2)}")
    
    elif operacao == 2:
        a = eval(input(f"\nDigite o escalar: "))
        k1 = eval(input(f"Digite o ket: "))
        print(f"O produto é: {multip_escalar(a, k1)}")
    
    elif operacao == 3:
        k1 = eval(input(f"\nDigite o ket: "))
        print(f"O respectivo bra é: {bra(k1)}")
    
    elif operacao == 4:
        k1 = eval(input(f"\nDigite o primeiro ket (esse será convertido em bra): "))
        k2 = eval(input(f"Digite o segundo ket: "))
        print(f"O produto interno é: {produto_interno(k1, k2)}")
    
    elif operacao == 5:
        k1 = eval(input(f"\nDigite o ket: "))
        print(f"O módulo desse ket é: {modulo(k1)}")
    
    elif operacao == 6:
        k1 = eval(input(f"\nDigite o ket: "))
        print(f"O ket normalizado é: {normalizado(k1)}")
    
    elif operacao == 7:
        k1 = eval(input(f"\nDigite o primeiro ket: "))
        k2 = eval(input(f"Digite o segundo ket (esse será convertido em bra): "))
        print(f"O produto externo é: {produto_externo(k1, k2)}")
    
    elif operacao == 8:
        operador = eval(input(f"\nDigite o operador: "))
        k1 = eval(input(f"Digite o ket: "))
        print(f"A aplicação desse operador nesse ket é: {operador_ket(k1, operador)}")
    
    elif operacao == 9:
        operador = eval(input(f"\nDigite o operador: "))
        k1 = eval(input(f"Digite o primeiro ket: "))
        k2 = eval(input(f"Digite o segundo ket: "))
        print(f"O elemento de matriz é: {elemento_matriz(k1, k2, operador)}")
    
    elif operacao == 10:
        operador = eval(input(f"\nDigite o operador: "))
        k1 = eval(input(f"Digite o ket: "))
        print(f"O valor esperado é: {valor_esperado(k1, operador)}")
    
    else:
        print(f"\nDigite um dos números indicados acima!")
        print("Reiniciando o menu...\n")
        continue

    resposta = input("\nDeseja realizar outra operação? (s/n)").strip().lower()

    if resposta != "s":
        print("\nObrigado por usar o calculador... Até mais!")
        continuar = False
```
